## 10 x 10 matricies
this weeks assignment asks us to write a program that generates 100 random integers 0 through 9

### Import Duties

we'll start with importing the 'random' python library 

```
import random
```

### Mega Maniacal Matrix Maker Madness
next we'll write a function that will create a new 10 x 10 matrix each time it's called. 

This will make it easy to create unique matricies at the stroke of a key.

Or, if we want to, we can store the matrix in a variable to reuse as we like.

```
# make a 10x10 matrix of random numbers 0-9
matrix_10x10 = lambda: [[random.randint(0,9) for rand_num in range (10)] for lst in range(10)]
matrix_10x10()
```

Yay! that's our **matrix**, we could make it _prettier_ though.

### Stringy Thingy

What we'll do next is convert the matrix into a string using function just for that purpose.

```
# Print them in a 10 by 10 matrix neatly arranged (one space between each number):

matrix_to_string = lambda some_matrix: '\n'.join([' '.join(str(num) + ' ' for num in lst) for lst in some_matrix])

print (matrix_to_string(matrix_10x10()), '\n')
```

That's a good looking matrix!

### @ Odds With Random Numbers
Now we can try to replace any odd number with @ and see how that looks.

We'll do that by writing another function.

```
# If the random number is an odd number, print "@" instead. :

odd_num_to_at = lambda: [['@' if num%2 != 0 else num for num in lst] for lst in matrix_10x10()]

print(matrix_to_string(odd_num_to_at()), '\n')
```

Definitely funky lookin'!

### It's All Adding Up Now!

Something more practical to do is to display the sum of all the numbers in each row.

You might have noticed something of a patteren because we're going to write another function to do just that!

```
# Calculate and Print the total of each row like the following (use "*" to separate the total from the numbers):

def sum_matrix_row(some_matrix):
  sum_row = [sum(lst) for lst in some_matrix if type(lst) is list]
  
  [some_matrix[i].append('* '+ str(sum_row[i])) for i in range (len(some_matrix[0]))]
    
    
  return some_matrix

print (matrix_to_string(sum_matrix_row(matrix_10x10())) , '\n')
```

Cool and useful to boot!
### Big Bright Bodacious Boundaries!
We can add some aesthetic flair by surrounding our matrix with twinkly little stars.

The next function will add a row of *'s to the top and bottom, and as the first and last element of each list.


```
# Optional: Surround the matrix with asterisks (*) like the following:
def beautiful_borders(some_matrix):
  border_row = '*' * (len(some_matrix) + 2)
  [(lst.append('*'), lst.insert(0, '*')) for lst in some_matrix]
  some_matrix.insert(0, border_row)
  some_matrix.append(border_row)

  return matrix_to_string(some_matrix)
  
print(beautiful_borders(matrix_10x10()), '\n')
```

Wow! Our little matrix looks like it's nestled in the night's sky. What a sight to behold!

### The Final Countdown
Last, but not least, is to display the sums of the columns as well as the rows.

We'll make a new function that will add up the columns and also call our previous function sum_matrix_row to provide the sum for the rows.

And we'll neatly (I couldn't get the spacing quite right) display them on either side of the matrix separted by *'s.

(those who are interested to know I know that there is a problem with spacing, the sums are 2 digits but the numbers in the matrix are only 1 digit wide, but I didn't want to mess with my matrix_to_string function too much, oh well...)

```
# Optional (challenging): Calculate and Print the total of each row and column like the following:

def sum_matrix_columns_rows(some_matrix):
  # print(matrix_to_string(some_matrix))
  sum_columns = [sum(lst) for lst in zip(*some_matrix)]
  some_matrix.append('*'*len(some_matrix))
  some_matrix.append(sum_columns)
  sum_matrix_row(some_matrix)
  return some_matrix


print (matrix_to_string(sum_matrix_columns_rows(matrix_10x10())))
```
